## D
Format all the column names to lowercase and include underscore between column names that consist of two words. For example, meanTemp should become mean_temp, and Max24hrPrep can become max_24hr_prep, and HighTemp becomes high_temp, etc. Reassign the DataFrame with the formatted column names to the same variable, clean_boston_data. Print or output the columns of the clean_boston_data DataFrame. 

```python
# Get the current names for the columns
current_names = clean_boston_data.columns.values.tolist()
# Create an empty array for the new names
new_names = []
# regex pattern to split the names by uppercase and hour format
pattern = re.compile(r'(?<=[a-zA-Z0-9])(?=[A-Z])|(?<=[A-Za-z])(?=\d{2})')
# Loop through the names and split them by uppercase and hour format, then join them again with an underscore between them
for col in current_names:
    split_col = pattern.split(col)
    split_col = [x.lower() for x in split_col]
    new_col = "_".join(split_col)
    new_names.append(new_col)
# Assign the new names
clean_boston_data.columns = new_names
```

## A
Use the LabelBinarizer() in the sklearn package to transform the ice_cream_flavor column in the student_data to dummy variables, then join these dummy variables to the student_data and drop the original ice_cream_flavor column. Reassign the resulting DataFrame to a variable called student_data_1. Print out the entire student_data_1 DataFrame 

```python
# Create the LabelBinarizer instance
lb = LabelBinarizer()
# Fit it with the ice_cream_flavor column
lb.fit(student_data["ice_cream_flavor"])
# Create the new dataframe from the previous dataframe
student_data_1 = student_data.copy()
# Create the dummy variables using the lb instance
dummies = pd.DataFrame(lb.transform(student_data["ice_cream_flavor"]))
# Get the names for the columns from the classes identified in the lb
dummies.columns = lb.classes_
# concat the student_data_1 dataset with the dummy variables
student_data_1 = pd.concat([student_data_1, dummies], axis=1)
# Drop the original ice_cream_flavor column
student_data_1 = student_data_1.drop(["ice_cream_flavor"], axis=1)
student_data_1
```

```python
# Creates the standard scaler instance and fits it with the scores
scaler = StandardScaler()
scores = student_data_1[["math", "reading", "science "]].values
scaler.fit(scores)
# scale the 3 scores: math, science, reading
scaled_scores = scaler.transform(scores)
# Creates a new dataframe from the scaled scores and adds the column name
std_scores = pd.DataFrame(scaled_scores, columns=["std_math", "std_reading", "std_science"])
```

Using the boston_1 DataFrame in question 1a, select the LowTemp, HighTemp, WarmestMin, ColdestHigh, AveMin, AveMax columns and use the “pipeline” functionality in sklearn to transform the selected data using the MinMaxScalar() and Imputer() classes.  With the Imputer() class, the missing data in each column should be imputed using the mean value for the column. Assign the resulting DataFrame to a variable called pipeline_data. Print the pipeline_data
NB: It appears the Imputer() class in sklean used in the course material is depreciated at this time. Alternatively, you can similarly use the SimpleImputer() class as follows:

```python
# Reads the boston_weather_1 file again
boston_data = pd.read_csv("boston_weather_1.csv")
# Keeps only select columns
boston_data = boston_data[["LowTemp", "HighTemp", "WarmestMin", "ColdestHigh", "AveMin", "AveMax"]]
boston_data
```

```python
# Create a new pipeline with a MinMax scaler, and a simple imputer
pipeline = make_pipeline(MinMaxScaler(), SimpleImputer(missing_values = np.nan, strategy = 'mean'))
```

```python
# Fits the pipeline with the boston data
pipeline.fit(boston_data)
```

```python
# Runs the pipeline with the boston data
pipeline_data = pipeline.transform(boston_data)
# Creates a dataframe from the data after the pipeline
pipeline_data = pd.concat([pd.DataFrame(), pd.DataFrame(pipeline_data)])
# Adds column names to the dataframe
pipeline_data.columns = ["LowTemp", "HighTemp", "WarmestMin", "ColdestHigh", "AveMin", "AveMax"]
pipeline_data
```

```python
# As the data was imputed, there should be no missing values
pipeline_data.isna().sum()
```
